# Opioid Arrests Analysis

Data comes from https://data.pa.gov/Opioid-Related/Opioid-Seizures-and-Arrests-CY-2013-Current-Quarte/wmgc-6qvd

This notebook is to quickly analyze the data on opioid related deaths in PA. 
The main question is: how have opioid arrest numbers change over the past 8 years.

## Steps
* Load in the Data
* Drop the year 2021 because of incomplete data
* Group the data by year and quarter, and then just year and sum the number of arrests
* Plot the data by year and quarter to see the change overtime
* Use percent change function to see how much arrests have changed

```python
arrests_df = pd.read_csv('../data/opioid_arrests.csv')
```

```python
arrests_df = arrests_df[arrests_df['Year'] != 2021]
```

```python
arrests_df.groupby(['Year','Qtr'])['Arrests'].sum().plot(title='Opioid Related Arrests by Quarter', use_index=True, figsize =(10,5))
plt.show()
```

```python
arrests_df.groupby('Year')['Arrests'].sum().plot(title='Opioid Related Arrests by Year 2013-2020', ylabel='Arrests')
plt.show()
```

```python
arrests_df.groupby('Year')['Arrests'].sum().pct_change().mean()
```

## Conclusions
* The number of arrests has increased, despite a brief decrease from 2016 to 2018
* This increase matches the increase in opioid deaths that we see at the federal level
* The number of arrests increased by 13.54% per year on average
* This data implies that opioid related arrests don't have much influence on opioid overdose deaths
