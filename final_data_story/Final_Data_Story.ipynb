# The Opioid Epidemic: How It Has Affected Connecticut and the Entire US

The opioid epidemic began in the late 1990s when pharmaceutical companies ensured medical professionals that opioid pain relievers were not addictive and safe to give to patients as pain relief medication. Doctors started prescribing opioids at an alarming rate. Eventually, it became clear that these medications were extremely addictive. People began to die from prescription opioid use. Not only were people overdosing on the prescription medications, but after patients were addicted to medication and their doctor stopped prescribing it, they turned to illegal opioids to feed their addiction. This led to even more overdoses from illegal opioids. 

According to drugabuse.gov, roughly 21 to 29 percent of patients who have been precribed opioids for chronic pain misuse them. Additionally, between 8 to 12 percent of people who are prescribed an opioid become addicted and 4 to 6 percent of those people transition to using heroin and their drug of choice. In fact, about 80% of people who first use heroin began by misusing prescription opiods. 

The opioid epidemic has created a public health crisis. Opioid misuse has lead to increased illegal drug use, increased opioid-related overdoses, as well as in increase in neonatal abstinence syndrome. Neonatal abstinence syndrome occurs when the fetus is exposed to drugs in the womb and is born addicted to that drug. 
(https://www.drugabuse.gov/drug-topics/opioids/opioid-overdose-crisis) 

The opioid epidemic is one that can affect a large variety of people regardless of age, gender, or socioeconomic status. Therefore, we chose to focus our data story on the state of Connecticut. According to census.gov, the median household income in 2019 in Connecticut was \\$78,444, which is \\$15,601 higher than the federal household income. Additionally, the percent of people living in poverty was 9.7% in 2019, while the national average was 11.4%. (https://www.census.gov/quickfacts/fact/table/CT,US/PST045219). From this data, we can see that Connecticut is a state that is above average in terms of socioeconomic status.  

Our data story evaluates how the opioid epidemic has affected the state of Connecticut by looking at the opioid prescription rates from 2014-2020 as well as rehab admissions and accidental opioid overdoses. Then, we look at some federal data on opioid prescribing rates and total opioid deaths. Lastly, we look at opioid-related arrests to evaluate how the opioid epidemic has transitioned from people misusing legally-prescribed opioids to illegal opioid use.

### Federal Opioid Prescribing Rates

Because we looked at opioid deaths at the federal level, we were also curious about national opioid prescribing rates. We found data on data.gov drawn from Department of Veterans Affairs medical centers across the country that recorded 2012 and 2018 opioid prescribing rates (https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018). 

For clarity, the prescribing rate is the total percentage of total prescriptions administered by statewide VA medical centers, according to USA Today (https://www.usatoday.com/story/news/politics/2018/01/11/va-publicly-releases-opiate-prescribing-rates-all-its-hospitals-first-time/1024895001/.
). 

First, we wanted to discover how Connecticut compares to other states. We found that in 2012, Connecticut was ranked as the state with the 4th lowest prescribing rate at 8.9%. In 2018, Connecticut was ranked 8th lowest with a 6.5% prescribing rate. Additionally, this data is consistent with the Connecticut state data from above that shows its prescribing rate has decreased over time. 



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))

However, because Connecticut was so low relative to other states, we were curious what the top states were for opioid prescribing rates.

The following table shows the top 5 states with the highest prescribing rates in 2012 in descending order: Oklahoma, Nevada, Oregon, Arkansas, and New Mexico. They were all above 23% of total prescriptions.



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))

The following table shows the top 5 states with the highest prescribing rates in 2018 in descending order: Idaho, Oklahoma, New Mexico, Oregon, and Montana. They were all above 12% of total prescriptions but far less than the rates from 2012. 



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))

From this analysis, we noticed that in 2012, the state with the highest opioid prescribing rate, Oklahoma, had a rate of 24.3% which is 2.73x that of Connecticut's. In 2018, the highest state was Idaho, with a rate of 15.0% which is 2.31x that of Connecticut's. In other words, Connecticut's opioid prescribing rate in each year was less than half of the state with the highest prescribing rate that year!

Furthermore, there is an apparent trend that prescribing rates have drastically lowered from 2012 to 2018 nationwide as well. Oklahoma was the highest in 2012 at a 24.3% prescribing rate whereas in 2018, Idaho was the highest with just 15.0%.

Based on our observations, it seems like states in the Midwest/West have the highest opioid prescribing rates. For example, Oklahoma, Oregon, and New Mexico were in the top 5 states for both 2012 and 2018. This made us wonder: Is this a national trend? <b> Does longitude have an effect on the opioid prescribing rates across the country? </b> 

We initially hypothesized that if there were to be such a relationship, the prescribing rate would increase as we move further West across the country.



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))

(*Note: We filtered out Alaska, Hawaii, Philippines, and Puerto Rico from the data to focus on the contiguous US)

After conducting correlation analysis between longitude and the 2012 opioid prescribing rate, we found the correlation to be -0.508 which is a <b> moderate negative relationship.</b> This is significant because there does seem to be a relationship where as we move East, the prescribing rate decreases. Conversely, as we move West, the prescribing rate increases, which aligns with our hypothesis!



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))

(*Note: We filtered out Alaska, Hawaii, Philippines, and Puerto Rico to focus on the contiguous US)

The relationship between longitude and the 2018 prescribing rate is similar with a correlation of -0.496. Again, this is further evidence that as we move further West, the opioid prescribing rate increases.

We speculate that some reasons causing this relationship are 1) more relaxed opioid prescription laws; and 2) the increased proportion of rural areas relative to urban areas in the West. 

Below is a visualization of this phenomenon on a map using the 2018 opioid prescribing rates since it is the most recent data. We plotted all the data points of medical centers in the contiguous US with a color scheme from dark purple to yellow. As the color gets lighter, the opioid prescribing rate, written as a proportion (e.g. 0.10 = 10%), increases.  



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))

From the map above, we can see a smattering of yellow and orange centers, which have higher opioid prescribing rates at 20-25%, in California and across the Midwest and South. There are many purple centers (<15% opioid prescribing rate) clustered around the Northeast. 

To better visualize the 2018 prescribing rate by state, we found the mean prescribing rate and colored in the map below.



(Sharon, [Source](https://catalog.data.gov/dataset/opioid-prescribing-rates-at-va-facilities-2012-2018))

On this map, the dark purple indicates a low mean opioid prescribing rate and as the color scheme gets lighter to yellow, th mean opioid prescribing rate increases. The number 0.050 is the proportion of opioid prescriptions (e.g. 5% prescribing rate).

Pinpointing Connecticut, we can see that it is dark purple, which makes sense since we already found their prescribing rate was low compared to the other states. Idaho, which has the highest 2018 prescribing rate at 15.0% is the lighest color, a pale yellow, which stands in contrast with the shades of orange and purple on the graph.

Overall, it is clear that there is a general trend of darker colors and lower opioid prescribing rates on the East and top Midwest while the South and the West have lighter colors and consequently, higher opioid prescribing rates.

### Opioid Arrest Data

Since we the group saw a clear increase in federal opioid overdose deaths but a decrease in federal opioid prescribing rates, we were interested in seeing how opioid related arrests changed in that time frame.

We believe that the decrease in legal opioid presciptions but the increase in overdose deaths could signify an increase in illegal opioid sales. Our logic allows us to hypothesize that opioid related arrests should also increase if illegal opioid sales are more prevalent. 

The group analyzed data from the state of Pennsylvania counting opioid arrests and seizures since 2013. The data was from data.pa.gov 

https://data.pa.gov/Opioid-Related/Opioid-Seizures-and-Arrests-CY-2013-Current-Quarte/wmgc-6qvd

By analyzing the data by year, we can see that opioid arrests have increased dramatically since 2013. The average number of arrests is increasing at a rate of 13.5% per year.



(Jonah [Source](https://data.pa.gov/Opioid-Related/Opioid-Seizures-and-Arrests-CY-2013-Current-Quarte/wmgc-6qvd))

This data is striking because it confirms our suspicions. While this data does not prove any causative relationship between increased opioid overdose deaths or decreased prescriptions, this correlation is enough to warrant closer investigation of the legal and illegal ways of obtaining opioids. 

## Summary of Findings



Throughout our analyses, we made several significant findings about the opioid epidemic. We found that:
* Connecticut opioid prescriptions have decreased from 2012 to 2020.
* Connecticut rehab admissions are increasing.
* Accidental opioid deaths have increased from 2012 to 2020.
* Federal overdose deaths increased from 2015-2020 and opioid deaths also increased in the same time fram eat a faster rate.
* Connecticut has a large amount of opioid overdose deaths compared to other states.
* Federal opioid prescribing rates for Connecticut are low compared to other states.
* Nationwide, opioid prescribing rates increase as we move from East to West on a map of the contiguous US.
* Pennsylvania opioid related arrests have increased from 2013 to 2020. 

Overall, opioid prescriptions are decreasing as a whole for both Connecticut and on a national level. While this is promising, opioid overdose deaths have increased nationally, with Connecticut having a large amount of deaths compared to other states. 

Since legal opioids prescribed from hospitals and medical professionals have decreased but deaths are increasing, this dichotomy indicates that there may be a growing use of illegal opioids nationwide. Even in a state like Connecticut that is above average in socioeconomic status, the opioid epidemic is not to be taken lightly, especially as we also found data that showed increasing rehab admissions in the state. Some questions to think about for the future are: How can we as a country combat underground opioids? Should we put more rules and regulations in place? What demographics are those most affected by opioid addiction, and help do we help them?

Also, if opioid arrests are increasing but overdose deaths are still increasing, how should we change the way we are combatting underground opioid distribution? Are there better ways to fight illegal distribution and prevent overdose deaths at the same time?

Therefore, in addition to the numbers from our datasets, we understand that the opioid crisis is not simply black and white. There are a variety of socioeconomic and political issues surrounding the epidemic affecting the government, Big Pharma companies, and of course, the people in our country whose lives have been negatively impacted by these drugs.
